```python
#Import dependencies
import pandas as pd
```

```python
#Read the weather_data.csv file
weather_path = "weather_data.csv"

#Create a dataframe from the weather_data.csv file
weather_df = pd.read_csv(weather_path)

#Reorder the dataframe columns
column_list = ['City ID', 'City', 'Country', 'Latitude', 'Longitude', 'Max Temperature', 'Humidity', 'Cloudiness', 'Wind Speed', 'Timestamp']
weather_df = weather_df[column_list]

#Rename columns
weather_df.rename(columns = {'Max Temperature': 'Max Temperature (°F)', 'Humidity': 'Humidity (%)', 'Cloudiness': 'Cloudiness (%)', 'Wind Speed': 'Wind Speed (MPH)'}, inplace = True)

#Print weather_df dataframe
weather_df.head()
```

```python
#Export the weather_df dataframe to html
    #https://pandas.pydata.org/pandas-docs/version/0.17.0/generated/pandas.DataFrame.to_html.html
weather_df.to_html('weather_data.html', index = False, header = True)
```
